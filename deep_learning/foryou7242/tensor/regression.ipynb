```python
# predict exam score
#hours score
#10      90
#9       80
#3       50
#2       30
#called training set
#regression 
#x y
#1 1
#2 2 
#3 3

# H(x) = Wx + b
#cost(loss) function how fit the line to our data
#H(x) - y
#(H(x) -y) ^2

data_set = [10, 9, 3, 2]
score_set = [90, 80, 50, 30]

w = tf.Variable(tf.random_normal([1], name='w'))
b = tf.Variable(tf.random_normal([1], name='b'))

hypothesis = data_set * w + b

print(hypothesis)
cost = tf.reduce_mean(tf.square(hypothesis-score_set))
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(cost)


sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(10000):
    sess.run(train)
    if step % 20== 0:
        print(step, sess.run(cost), sess.run(w), sess.run(b))
        we = sess.run(w)
        bi = sess.run(b)
x0 = np.arange(-2, 2.5, 0.25)
        
plt.xlim([-2, 50])
plt.ylim([-2, 50])
plt.xlabel('x')     
plt.ylabel('y')
y = x0 * we + bi
        
plt.plot(x0,y)
        
        
        
        
```
