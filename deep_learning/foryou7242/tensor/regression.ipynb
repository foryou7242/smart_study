```python
#place holders
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
w = tf.Variable(tf.random_normal([1], name='w'))
b = tf.Variable(tf.random_normal([1], name='b'))

hypothesis = x * w + b


cost = tf.reduce_mean(tf.square(hypothesis-y))
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(cost)
sess = tf.Session()
sess.run(tf.global_variables_initializer())
size = 3000
ys = []
for step in range(size):
    cost_val, W_val, b_val,_ = sess.run([cost, w,b, train], feed_dict={x:[1.0,2.0,3.0], y:[3.1,4.1,5.1]})
    ys.append(cost_val)
    print(f'{W_val},{b_val}, {cost_val}')

plt.xlim([-2, size])
plt.ylim([-2, 10])
plt.xlabel('x')     
plt.ylabel('y')
#y = x0 * we + bi
print(sess.run(w))
print(sess.run(b))
print(sess.run(hypothesis, {x:[1.0,2.0,3.0]}))

plt.plot(ys)

#print(sess.run(hypothesis, feed_dict={x:[1.2]}))

        
```
