```python
from collections import deque
```

What is the yield keyword?

```python
def num_gen():
    n = 0
    while n < 10000000:
        yield n
        n += 1
```

```python
gen = num_gen()
```

```python
next(gen)
```

```python
def search(lines, pattern, history=5):
    previous_lines = deque(maxlen=history)
    for line in lines:
        previous_lines.append(line)
        if pattern in line:
            yield previous_lines
        
```

```python
with open("03_keeping_the_last_n_items_some_file.py") as f:
    cont = f.read()
```

```python
def show_functions(content):
    content = content.split('\n')
    content.reverse()
    search_gen = search(content, pattern="def ", history=5)
    while True:
        try:
            def_lines = next(search_gen)
            print("\n".join(reversed(def_lines)))
            print("\n...\n\n")
        except StopIteration:
            return
```

```python
show_functions(cont)
```
