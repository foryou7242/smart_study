```python
with open('/etc/passwd') as f:
    try:
        while True:
            line = next(f)
            print(line, end ='')
    except StopIteration:
        pass
    
```

```python
with open('/etc/passwd') as f:
    while True:
        line = next(f, None)
        if line == None:
            break
        print(line, end ='')
```

```python
import numpy as np
class Cards():
    def __init__(self):
        self.cards = np.random.choice(60000,20)
        self.idx = 0
    def __iter__(self):
        return self
    def __next__(self): 
        self.idx += 1
        if self.idx == 20:
            self.idx = 0
        return self
    def __str__(self):
        return f'{self.cards[self.idx]}'
cards = Cards()
print(cards.cards)

it = iter(cards)

for i in range(1000):
    print(it)
    it = next(it, None)
    
    if it == None:
        break;
    
    
```
